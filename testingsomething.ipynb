```python
def getfreeperiod(x):
    for i in range(1,8):
        if class5a[x][i]=="nil":
            return (x,i)
    return (0,0)
```

```python
def isteacherfree(teacheris,day,period):
    if teachertimetable[teacheris][day][period]=="nil":
        return True
    else:
        return False
```

```python
def issubjectonday(subject,day):
    for i in range(1,8):
        if class5a[day][i]==subject:
            return True
    return False
```
